```python
# Third party libraries
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt

# Sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# TensorFlow
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical


# ------------------ Data Loading ------------------
# Load the training CSV file
train = pd.read_csv('train.csv')
print("Training data preview:")
print(train.head())


# ------------------ Data Visualization ------------------
def show_sample_images(df, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(10, 2))
    pixel_cols = [col for col in df.columns if col != 'label']
    for i, ax in enumerate(axes):
        image = df.iloc[i][pixel_cols].values.astype(np.float32)
        image = image.reshape(28, 28)
        ax.imshow(image, cmap='gray')
        ax.axis('off')
        ax.set_title(f"Label: {df.iloc[i]['label']}")
    plt.tight_layout()
    plt.show()


# Display a few sample images from the training data
show_sample_images(train, num_images=5)


# ------------------ Data Preprocessing ------------------
# Separate features and labels
X = train.drop("label", axis=1).values.astype(np.float32)
y = train["label"].values

# Normalize pixel values to [0, 1]
X = X / 255.0

# One-hot encode the labels
num_classes = len(np.unique(y))
y_cat = to_categorical(y, num_classes)

# Perform an 80/20 split for training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# Reshape the data for the CNN (28x28 images with 1 channel)
X_train = X_train.reshape(-1, 28, 28, 1)
X_val = X_val.reshape(-1, 28, 28, 1)


# ------------------ Model Definition ------------------
model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()


# ------------------ Model Training ------------------
history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=10, 
            batch_size=64
        )


# ------------------ Training History Visualization ------------------
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], marker='o', label='Train Accuracy')
plt.plot(history.history['val_accuracy'], marker='o', label='Validation Accuracy')
plt.title("Accuracy over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], marker='o', label='Train Loss')
plt.plot(history.history['val_loss'], marker='o', label='Validation Loss')
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()


# ------------------ Evaluation and CSV Output ------------------
# Predict on the validation set
y_val_pred_probs = model.predict(X_val)
y_val_pred = np.argmax(y_val_pred_probs, axis=1)
y_val_true = np.argmax(y_val, axis=1)

# Save predictions vs truths to a CSV file
results_df = pd.DataFrame({
    "Truth": y_val_true,
    "Prediction": y_val_pred
})
results_df.to_csv("predictions_vs_truth.csv", index=False)
print("Predictions vs Truth saved to predictions_vs_truth.csv")

# Print the confusion matrix and classification report
cm = confusion_matrix(y_val_true, y_val_pred)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_val_true, y_val_pred))

# Plot the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[str(i) for i in range(num_classes)])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix Display")
plt.show()


# ------------------ Display Example Predictions ------------------
num_examples = 5
fig, axes = plt.subplots(1, num_examples, figsize=(15, 3))
for i in range(num_examples):
    ax = axes[i]
    image = X_val[i].reshape(28, 28)
    ax.imshow(image, cmap='gray')
    ax.axis('off')
    ax.set_title(f"Truth: {y_val_true[i]}\nPred: {y_val_pred[i]}")
plt.suptitle("Example Predictions (Truth vs Prediction)")
plt.tight_layout()
plt.show()


# ------------------ Visualizing Feature Maps for All Convolutional Layers and All Numbers ------------------

def plot_activation_maps(activation, layer_name, digit):
    """
    Plot all feature maps for a given activation.
    activation: numpy array of shape (1, height, width, channels)
    """
    n_features = activation.shape[-1]
    # Determine grid size (try to form a square-ish layout)
    n_cols = int(np.ceil(np.sqrt(n_features)))
    n_rows = int(np.ceil(n_features / n_cols))
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*1.5, n_rows*1.5))
    fig.suptitle(f"Digit {digit} - {layer_name} Feature Maps", fontsize=16)
    
    # Flatten axes for easy iteration
    axes = axes.flatten() if n_features > 1 else [axes]
    for i in range(n_features):
        ax = axes[i]
        ax.imshow(activation[0, :, :, i], cmap='viridis')
        ax.axis('off')
    # Remove any extra subplots
    for j in range(n_features, len(axes)):
        axes[j].axis('off')
    plt.tight_layout()
    plt.show()

# Get all convolutional layers from the model
conv_layers = [layer for layer in model.layers if isinstance(layer, Conv2D)]

# Build an activation model that outputs the activations from all convolutional layers.
activation_model = Model(inputs=model.layers[0].input, outputs=[layer.output for layer in conv_layers])

# Get the true labels for the validation set (as integers)
y_val_labels = np.argmax(y_val, axis=1)

# Loop over each digit from 0 to 9
for digit in range(10):
    indices = np.where(y_val_labels == digit)[0]
    if len(indices) == 0:
        continue  # if there is no sample for this digit, skip
    idx = indices[0]  # take the first sample for that digit
    sample_image = X_val[idx].reshape(1, 28, 28, 1)
    # Get the activations for this sample from all conv layers
    activations = activation_model.predict(sample_image)
    print(f"\nVisualizing feature maps for digit {digit} (sample index {idx}):")
    for i, activation in enumerate(activations):
        layer_name = conv_layers[i].name
        plot_activation_maps(activation, layer_name, digit)
```

```python
from tensorflow.keras.preprocessing import image
import cv2

def predict_digit_from_image(img_path):
    """
    Predict the digit from a custom image.
    
    Parameters:
        img_path (str): Path to the image file.
    
    Returns:
        int: Predicted digit label.
    """
    # Load the image in grayscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # Resize to 28x28 (as used in training)
    img_resized = cv2.resize(img, (28, 28))

    # Invert colors if background is white (optional, depending on image)
    if np.mean(img_resized) > 127:
        img_resized = 255 - img_resized

    # Normalize pixel values
    img_normalized = img_resized.astype(np.float32) / 255.0

    # Reshape to (1, 28, 28, 1) for model prediction
    img_input = img_normalized.reshape(1, 28, 28, 1)

    # Predict using the trained model
    prediction = model.predict(img_input)
    predicted_label = np.argmax(prediction)

    # Show the image and prediction
    plt.imshow(img_resized, cmap='gray')
    plt.title(f"Predicted Digit: {predicted_label}")
    plt.axis('off')
    plt.show()

    return predicted_label


# Example usage 1
predict_digit_from_image(r"C:\Users\Siva\OneDrive\Pictures\Screenshots\Screenshot 2025-04-17 224354.png")
# Example usage 2
predict_digit_from_image(r"C:\Users\Siva\Downloads\5.png")
```
